# Cross Correlation and DFT

```python
def crossCorrelation(signal1, signal2, verbose = True):
    
    time1 = time.time()
    N = len(signal1)
    
    res = np.zeros(N)
    for n in range(N):
        
        for m in range(N):
            res[n] += signal1[m] * signal2[(m+n)%N]
    time2 = time.time()
    if verbose == True:
        print('Cross correlation without DFT took {:.3f} ms'.format((time2-time1)*1000.0))
    
    
    return res
```

```python
crCor = crossCorrelation(sensor1[:100], sensor2[:100])
```

```python
time = [i for i in range(len(crCor))]
plt.plot(time, crCor)
```
